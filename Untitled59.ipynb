**PSD ANALYSIS**

```python
import mne
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from mne.time_frequency import psd_array_welch
```

```python
# Function to read and preprocess data
def read_data_psd(file_path):
    datax = mne.io.read_raw_edf(file_path, preload=True)
    datax.set_eeg_reference()
    datax.filter(l_freq=1, h_freq=45)

    # Split the data into two equal parts (rest and task)
    total_time = datax.times[-1]
    mid_time = total_time / 2

    rest_data = datax.copy().crop(tmin=0, tmax=mid_time)
    task_data = datax.copy().crop(tmin=mid_time, tmax=total_time)

    return rest_data, task_data

# Function to calculate PSD
def calculate_psd(raw_data, fmin=1, fmax=100):
    sfreq = raw_data.info['sfreq']
    psd, freqs = mne.time_frequency.psd_array_welch(raw_data.get_data(), sfreq=sfreq, fmin=fmin, fmax=fmax,
                                                    n_fft=1024, n_overlap=512, verbose=False)
    return psd, freqs

# Function to extract band power
def get_band_power(psd, freqs):
    delta = np.mean(psd[:, (freqs >= 1) & (freqs < 4)], axis=1)
    theta = np.mean(psd[:, (freqs >= 4) & (freqs < 8)], axis=1)
    alpha = np.mean(psd[:, (freqs >= 8) & (freqs < 12)], axis=1)
    beta = np.mean(psd[:, (freqs >= 12) & (freqs < 30)], axis=1)
    gamma = np.mean(psd[:, (freqs >= 30) & (freqs <= 100)], axis=1)
    return np.stack((delta, theta, alpha, beta, gamma), axis=0)

# Function to analyze PSD
def analyze_psd(file_paths):
    rest_psds = []
    task_psds = []

    for file_path in file_paths:
        rest_data, task_data = read_data_psd(file_path)

        rest_psd, freqs = calculate_psd(rest_data)
        task_psd, _ = calculate_psd(task_data)

        rest_band_power = get_band_power(rest_psd, freqs)
        task_band_power = get_band_power(task_psd, freqs)

        print(f"File: {file_path}")
        print(f"Rest PSD shape: {rest_band_power.shape}")
        print(f"Task PSD shape: {task_band_power.shape}")

        rest_psds.append(rest_band_power)
        task_psds.append(task_band_power)

    # Convert lists to arrays for easier manipulation and analysis
    rest_psds = np.array(rest_psds)
    task_psds = np.array(task_psds)

    print(f"Final Rest PSD shape: {rest_psds.shape}")
    print(f"Final Task PSD shape: {task_psds.shape}")

    return rest_psds, task_psds

# Perform PSD analysis
rest_psds, task_psds = analyze_psd(all_file_path)
```

```python
# Visualize the results
def plot_psd(rest_psds, task_psds):
    bands = ['Delta', 'Theta', 'Alpha', 'Beta', 'Gamma']

    plt.figure(figsize=(12, 6))
    x = np.arange(len(bands))
    width = 0.35

    rest_mean = rest_psds.mean(axis=(0, 2))
    task_mean = task_psds.mean(axis=(0, 2))
    rest_std = rest_psds.std(axis=(0, 2))
    task_std = task_psds.std(axis=(0, 2))

    plt.bar(x - width/2, rest_mean, width, label='Rest', yerr=rest_std)
    plt.bar(x + width/2, task_mean, width, label='Task', yerr=task_std)

    plt.xlabel('Frequency Bands')
    plt.ylabel('Power (V/Hz)')
    plt.title('Average Power in EEG Frequency Bands: Rest vs Task')
    plt.xticks(x, bands)
    plt.legend()
    plt.show()

plot_psd(rest_psds, task_psds)
```

```python
def compare_states(rest_psds, task_psds):
    bands = ['Delta', 'Theta', 'Alpha', 'Beta', 'Gamma']
    for i, band in enumerate(bands):
        t_stat, p_value = stats.ttest_rel(rest_psds[:, i].mean(axis=1), task_psds[:, i].mean(axis=1))
        print(f"{band} band:")
        print(f"  Rest mean: {rest_psds[:, i].mean()}, Task mean: {task_psds[:, i].mean()}")
        print(f"  t-statistic: {t_stat}, p-value: {p_value}")
        print()

compare_states(rest_psds, task_psds)
```

**EEGNet Implementation**

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, DepthwiseConv2D, Dropout, Dense, Flatten, AveragePooling2D, BatchNormalization, Activation, SeparableConv2D, Reshape, Permute
from tensorflow.keras.models import Model

def EEGNet(nb_classes, Chans=21, Samples=12500, dropoutRate=0.5, kernLength=64, F1=8, D=2, F2=16, norm_rate=0.25):
    input1 = Input(shape=(Samples, Chans))

    # Reshape the input to (Samples, Chans, 1)
    block1 = Reshape((Samples, Chans, 1))(input1)

    # Transpose to (Chans, Samples, 1)
    block1 = Permute((2, 1, 3))(block1)

    block1 = Conv2D(F1, (1, kernLength), padding='same', use_bias=False)(block1)
    block1 = BatchNormalization()(block1)
    block1 = DepthwiseConv2D((Chans, 1), use_bias=False, depth_multiplier=D, depthwise_constraint=tf.keras.constraints.max_norm(1.))(block1)
    block1 = BatchNormalization()(block1)
    block1 = Activation('elu')(block1)
    block1 = AveragePooling2D((1, 4))(block1)
    block1 = Dropout(dropoutRate)(block1)

    block2 = SeparableConv2D(F2, (1, 16), use_bias=False, padding='same')(block1)
    block2 = BatchNormalization()(block2)
    block2 = Activation('elu')(block2)
    block2 = AveragePooling2D((1, 8))(block2)
    block2 = Dropout(dropoutRate)(block2)

    flatten = Flatten(name='flatten')(block2)
    dense = Dense(nb_classes, name='dense', kernel_constraint=tf.keras.constraints.max_norm(norm_rate))(flatten)
    softmax = Activation('softmax', name='softmax')(dense)

    return Model(inputs=input1, outputs=softmax)

# Usage
EEGNetmodel = EEGNet(nb_classes=2, Chans=21, Samples=12500)
EEGNetmodel.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print model summary to verify the input shape
EEGNetmodel.summary()
```

```python
from sklearn.model_selection import GroupKFold,LeaveOneGroupOut
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
gkf=GroupKFold()
```

```python
gkf = GroupKFold(n_splits=5)  # Adjust the number of splits as needed
metrics = {
    'accuracy': [],
    'precision': [],
    'recall': [],
    'f1_score': []
}

for train_index, val_index in gkf.split(data_array, label_array, groups=group_array):
    train_features, train_labels = data_array[train_index], label_array[train_index]
    val_features, val_labels = data_array[val_index], label_array[val_index]

    # Normalize each channel separately
    for i in range(train_features.shape[2]):
        scaler = StandardScaler()
        train_features[:, :, i] = scaler.fit_transform(train_features[:, :, i])
        val_features[:, :, i] = scaler.transform(val_features[:, :, i])

    EEGNetmodel = EEGNet(nb_classes=2, Chans=21, Samples=12500)
    EEGNetmodel.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Convert labels to categorical
    train_labels_cat = tf.keras.utils.to_categorical(train_labels, num_classes=2)
    val_labels_cat = tf.keras.utils.to_categorical(val_labels, num_classes=2)

    EEGNetmodel.fit(train_features, train_labels_cat, epochs=50, batch_size=32, validation_data=(val_features, val_labels_cat), verbose=1)

    # Predict on validation set
    val_pred = EEGNetmodel.predict(val_features)
    val_pred_classes = np.argmax(val_pred, axis=1)

    # Calculate metrics
    metrics['accuracy'].append(accuracy_score(val_labels, val_pred_classes))
    metrics['precision'].append(precision_score(val_labels, val_pred_classes, average='weighted'))
    metrics['recall'].append(recall_score(val_labels, val_pred_classes, average='weighted'))
    metrics['f1_score'].append(f1_score(val_labels, val_pred_classes, average='weighted'))

# Print results
for metric, values in metrics.items():
    print(f"{metric.capitalize()}:")
    print(f"  Mean: {np.mean(values):.4f}")
    print(f"  Std Dev: {np.std(values):.4f}")
    print(f"  Values: {values}")
    print()
```

**TSCeption Implementation**

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, Activation, AveragePooling2D, Concatenate, DepthwiseConv2D, Flatten, Dense, Dropout
from sklearn.model_selection import GroupKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def TSCeption(nb_classes, Chans=21, Samples=12500, dropoutRate=0.5, norm_rate=0.25):
    input1 = Input(shape=(Chans, Samples, 1))

    # Temporal Convolutions
    convs = []
    for kernel_size in [3, 5, 7]:
        conv = Conv2D(32, (1, kernel_size), padding='same', use_bias=False)(input1)
        conv = BatchNormalization()(conv)
        conv = Activation('elu')(conv)
        conv = AveragePooling2D((1, 4))(conv)
        convs.append(conv)

    temporal_layer = Concatenate(axis=-1)(convs)

    # Depthwise Convolution
    depthwise = DepthwiseConv2D((Chans, 1), use_bias=False, depth_multiplier=2, depthwise_constraint=tf.keras.constraints.max_norm(1.))(temporal_layer)
    depthwise = BatchNormalization()(depthwise)
    depthwise = Activation('elu')(depthwise)
    depthwise = AveragePooling2D((1, 8))(depthwise)
    depthwise = Dropout(dropoutRate)(depthwise)

    flatten = Flatten(name='flatten')(depthwise)
    dense = Dense(nb_classes, name='dense', kernel_constraint=tf.keras.constraints.max_norm(norm_rate))(flatten)
    softmax = Activation('softmax', name='softmax')(dense)

    return Model(inputs=input1, outputs=softmax)

TSCmodel = TSCeption(nb_classes=2, Chans=21, Samples=12500)
```

```python
gkf = GroupKFold(n_splits=5)  # Adjust the number of splits as needed
metrics = {
    'accuracy': [],
    'precision': [],
    'recall': [],
    'f1_score': []
}

data_array = data_array.reshape(data_array.shape[0], 21, 12500, 1)

for train_index, val_index in gkf.split(data_array, label_array, groups=group_array):
    train_features, train_labels = data_array[train_index], label_array[train_index]
    val_features, val_labels = data_array[val_index], label_array[val_index]

    # Normalize each channel separately
    for i in range(train_features.shape[1]):  # Iterate over channels
      scaler = StandardScaler()
      train_features[:, i, :, 0] = scaler.fit_transform(train_features[:, i, :, 0])
      val_features[:, i, :, 0] = scaler.transform(val_features[:, i, :, 0])

    TSCmodel = TSCeption(nb_classes=2, Chans=21, Samples=12500)
    TSCmodel.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Convert labels to categorical
    train_labels_cat = tf.keras.utils.to_categorical(train_labels, num_classes=2)
    val_labels_cat = tf.keras.utils.to_categorical(val_labels, num_classes=2)

    TSCmodel.fit(train_features, train_labels_cat, epochs=50, batch_size=4, validation_data=(val_features, val_labels_cat), verbose=1)

    # Predict on validation set
    val_pred = TSCmodel.predict(val_features)
    val_pred_classes = np.argmax(val_pred, axis=1)

    # Calculate metrics
    metrics['accuracy'].append(accuracy_score(val_labels, val_pred_classes))
    metrics['precision'].append(precision_score(val_labels, val_pred_classes, average='weighted'))
    metrics['recall'].append(recall_score(val_labels, val_pred_classes, average='weighted'))
    metrics['f1_score'].append(f1_score(val_labels, val_pred_classes, average='weighted'))

# Print results
for metric, values in metrics.items():
    print(f"{metric.capitalize()}:")
    print(f"  Mean: {np.mean(values):.4f}")
    print(f"  Std Dev: {np.std(values):.4f}")
    print(f"  Values: {values}")
    print()
```
